# Introduction and Business Proposal

#### Research shows that many neighborhoods and suburbs in Chicago are stratified by income. As someone who lives in Chicago most of the time, I have observed this myself. For a moment, you may be walking in a neighborhood full of abandoned houses and people struggling to make ends meet, and some neighborhoods near you may live in mansions. In order to better study the economic difficulties faced by different neighborhoods/communities in Chicago, the University of Illinois at Chicago (UIC) has developed a "difficulty index." The economic hardship score is the average of six standardized variables: unemployment, education, per capita income level, poverty, crowded housing, and dependence.


#### It is conceivable that there are many factors that may affect the difficulty index of a block, such as its size, ethnic composition, and age. However, here we will explore how the type of site in a given block affects the difficulty index. The goal is to enable city officials and developers to determine or predict how different locations may affect the community’s economy. Using this information, developers can also more effectively decide which venues should be built or demolished to benefit the community.
